```python
#read dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
df=pd.read_csv(r"c:\datascience\projects\Vandhebharath_Eda\vande_bharath.csv.csv")
```

```python
# Check if there is any missing values or outliers in the dataset
missing_values = df.isnull().sum()
print(missing_values)
```

```python
df['Average occupancy']
```

```python
# Replace missing values or outliers with any mean or interpolated values
for i in df:
    if df[i].dtype!=object and df[i].dtype!='string':
        df[i]=df[i].interpolate(method='polynomial',order=5)
```

```python
df=df.dropna()
```

```python
df.info()
```

1. What is the distribution of the number of cars in trains?

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['No. of Cars'], bins=10, kde=True)
plt.title('Distribution of Number of Cars in Trains')
plt.xlabel('Number of Cars')
plt.ylabel('Frequency')
plt.show()
```

2.How do train operates frequency distribution?

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Frequency', data=df, palette='Set2')
plt.title('Frequency of Train Operations')
plt.xlabel('Frequency')
plt.ylabel('Count')
plt.show()
```

```python
df['Average Occupancy (%)'] = pd.to_numeric(df['Average occupancy'].str.replace('%', ''))

plt.figure(figsize=(8, 5))
sns.histplot(df['Average Occupancy (%)'], bins=10, color='blue', kde=True)
plt.title('Distribution of Average Occupancy Rates')
plt.xlabel('Average Occupancy (%)')
plt.ylabel('Frequency')
plt.show()
```

What is the number of  cars in train operators?

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Operator', y='No. of Cars', data=df, palette='pastel')
plt.title('Number of Cars by Train Operator')
plt.xlabel('Operator')
plt.ylabel('Number of Cars')
plt.xticks(rotation=45)
plt.show()
```

Which are the most common inauguration years for trains?

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Inauguration', order=df['Inauguration'].value_counts().index)
plt.title("Most Common Inauguration Years for Trains")
plt.xlabel("Inauguration Year")
plt.ylabel("Count")
plt.show()
```

What is the relationship between train speed and average occupancy?

```python
plt.figure(figsize=(10, 6))
sns.pointplot(data=df, x='Speed', y='Average occupancy')
plt.title("Speed vs. Average Occupancy")
plt.xlabel("Speed")
plt.ylabel("Average Occupancy")
plt.show()
```

Which operators manage the longest distance routes?

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Operator', y='Distance', estimator=np.max)
plt.title("Longest Distance Routes by Operator")
plt.xlabel("Distance")
plt.ylabel("Operator")
plt.show()
```

2. What are the most common train operators?

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Operator', order=df['Operator'].value_counts().index)
plt.title("Most Common Train Operators")
plt.xlabel("Frequency")
plt.ylabel("Operator")
plt.show()
```

What is the relationship between the distance and travel time of trains?

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Distance', y='Travel Time')
plt.title("Distance vs. Travel Time")
plt.xlabel("Distance")
plt.ylabel("Travel Time")
plt.show()
```

What is the average number of cars per operator?

```python
avg_cars_per_operator = df.groupby('Operator')['No. of Cars'].mean().sort_values()
plt.figure(figsize=(10, 6))
avg_cars_per_operator.plot(kind='barh', color='skyblue')
plt.title("Average Number of Cars per Operator")
plt.xlabel("Average Number of Cars")
plt.ylabel("Operator")
plt.show()
```

What is the distribution of travel times?

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Travel Time'], kde=True, bins=15)
plt.title("Distribution of Travel Times")
plt.xlabel("Travel Time")
plt.ylabel("Frequency")
plt.show()
```
